# Ejercicio 3
Generar la distribuci√≥n condicional para Brown Corpus e imprimir la tabla de frecuencia de las siguientes palabras:
* love
* hate
* speak
* control
* fear
* great
* president

En los g√©neros:
* news
* romance
* humor

```python
cfd = nltk.ConditionalFreqDist(
    (genre, word)
    for genre in brown.categories()
    for word in brown.words(categories=genre))
genres = ['news', 'romance', 'humor']
modals = ['love', 'hate', 'speak', 'control', 'fool', 'great', 'president']
cfd.tabulate(conditions=genres, samples=modals)
```

# Ejercicio 4
Imprimir los synsets de las palabras:
* computer
* machine
* car
* sandwich

```python
print("Synsets de computer: \n", wn.synsets('computer'))
print("Synsets de machine: \n", wn.synsets('machine'))
print("Synsets de car: \n", wn.synsets('car'))
print("Synsets de sandwich: \n", wn.synsets('sandwich'))
```

Imprimir la similitud entre: 
* computer y machine
* car y machine
* car y computer
* computer y sandwich

```python
# Guardamos los synsets de cada palabra
cm = wn.synset('computer.n.01')
ma = wn.synset('machine.n.01')
c = wn.synset('car.n.01')
sw = wn.synset('sandwich.n.01')

# Imprimimos las similitudes
print("Similitud entre computer y machine: ", cm.path_similarity(ma))
print("Similitud entre car y machine: ", c.path_similarity(ma))
print("Similitud entre car y computer: ", c.path_similarity(cm))
print("Similitud entre computer y sandwich: ", cm.path_similarity(sw))
```
