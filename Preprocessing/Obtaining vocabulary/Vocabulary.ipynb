# Getting the vocabulary from an HTML file

### Imports

```python
import nltk
from nltk.corpus import PlaintextCorpusReader
from bs4 import BeautifulSoup
```

### Reading the file and deleting the html tags

```python
f = open('e961024.htm', encoding = 'utf-8')
text = f.read()
f.close()

soup = BeautifulSoup(text, 'lxml')
words = soup.get_text()

print(words)
```

### Tokenizing the string we have

```python
tokens_nltk = nltk.word_tokenize(words)

print(tokens_nltk)
```

```python
tokens_py = words.split()

print(tokens_py)
```

### Eliminating the stop words and the punctuation signs

```python
# File from https://github.com/Alir3z4/stop-words/blob/master/spanish.txt
f = open('spanish_stop_words.txt', 'r')
stop_words = f.read()
f.close()

print(stop_words)
```

```python
f = open('spanish_punctuation signs.txt', 'r')
punctuation_signs = f.read()
f.close()

print(punctuation_signs)
```

```python
vocabulary = []

# If it is not a stop word nor a punctuation sign, it is part of the vocabulary

for word in tokens_nltk:
    if(word not in stop_words and word not in punctuation_signs):
        vocabulary.append(word)

print(vocabulary)
```

```python
# We write the words in lowercase
vocabulary = [word.lower() for word in vocabulary]
vocabulary.sort()

print(vocabulary)
```

```python
f = open('voc_html.txt', 'w')

for word in vocabulary:
    f.write(word + "\n")

f.close()
```
